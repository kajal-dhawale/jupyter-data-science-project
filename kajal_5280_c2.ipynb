```python
def extract_numeric(value):
    if pd.isna(value):
        return np.nan
    return float(''.join(c for c in str(value) if c.isdigit()))
df['Study Hours'] = df['Study Hours'].apply(extract_numeric)
df['Sleep'] = df['Sleep'].apply(extract_numeric)
df['Screen Time'] = df['Screen Time'].apply(extract_numeric)
df['Physical Activity'] = df['Physical Activity'].apply(extract_numeric)

df[['Study Hours','Sleep','Screen Time','Physical Activity']].head()

```

```python
df = df.fillna(df.mean(numeric_only=True))
for col in df.select_dtypes(include='object'):
    df[col] = df[col].fillna(df[col].mode()[0])
df.fillna(...)
df.head()
```

```python
# Standardize categorical columns
df['Diet'] = df['Diet'].str.lower().str.strip()
df['Living'] = df['Living'].str.lower().str.strip()
df['Extracurricular'] = df['Extracurricular'].str.lower().str.strip()
# SHOW OUTPUT
df[['Diet', 'Living', 'Extracurricular']].head()
```

```python
df['Performance'] = pd.to_numeric(df['Performance'], errors='coerce')
df['Performance'] = df['Performance'].fillna(df['Performance'].mean())
def performance_category(score):
    if score >= 8:
        return "High"
    elif score >= 5:
        return "Medium"
    else:
        return "Low"
df['Performance Category'] = df['Performance'].apply(performance_category)
df[['Performance', 'Performance Category']].head()
```

```python
df['Date'] = pd.to_datetime(
    df['Date'],
    errors='coerce',     
    format='mixed'       
)
df['Date'] = df['Date'].fillna(df['Date'].mode()[0])
df.dtypes
df['Date'].head()
```

```python
# Create Total Lifestyle Load column
df['Total Lifestyle Load'] = df['Study Hours'] + df['Screen Time']
# SHOW OUTPUT
df[['Study Hours', 'Screen Time', 'Total Lifestyle Load']].head()
```

```python
# Create Healthy Lifestyle column
df['Healthy Lifestyle'] = np.where(
    (df['Sleep'] >= 7) & (df['Physical Activity'] >= 60),
    'Healthy', 'Unhealthy'
)
# SHOW OUTPUT
df[['Sleep', 'Physical Activity', 'Healthy Lifestyle']].head()
```

```python
# Create Screen Category column
df['Screen Category'] = pd.cut(
    df['Screen Time'],
    bins=[0, 3, 6, 24],
    labels=['Low', 'Medium', 'High']
)
# SHOW OUTPUT
df[['Screen Time', 'Screen Category']].head()

```

```python
# Create Sleep Quality column
df['Sleep Quality'] = pd.cut(
    df['Sleep'],
    bins=[0, 5, 7, 24],
    labels=['Poor', 'Average', 'Good']
)

# SHOW OUTPUT
df[['Sleep', 'Sleep Quality']].head()
```

```python
df['Academic Risk'] = np.where(
    df['Performance Category']=='Low',
    'At Risk','Safe'
)

df[['Total Lifestyle Load','Healthy Lifestyle','Screen Category','Sleep Quality','Academic Risk']].head()
```
